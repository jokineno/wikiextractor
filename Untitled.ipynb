```python
import json 
import glob
import os 
```

```python
def get_all_files():
    return glob.glob("./text/*/wiki*")
```

```python
def title_to_id(article_json):
    title = article_json["title"]
    title_id = article_json["id"]
    return {title:title_id}

def read_file(path):
    with open(path, "r") as f:
        data = f.read()
    return data

def build_title_to_id_map(verbose=True):
    files = get_all_files()
    title2idmap = {}
    id2titlemap = {}
    for i, wiki_file in enumerate(files):
        if verbose and i % 100 == 0:
            print(f"Handled {i}/{len(files)}")
        articles = read_file(wiki_file).splitlines()
        for raw_article in articles:
            article = json.loads(raw_article)
            title2id = {article["title"]:article["id"]}
            id2title = {article["id"]:article["title"]}
            title2idmap.update(title2id)
            id2titlemap.update(id2title)
            
    return title2idmap, id2titlemap
```

```python
title2idmap, id2titlemap = build_title_to_id_map()
```
